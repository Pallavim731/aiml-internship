```python
import pandas as pd
```

```python
data = {
    "State":["Karnataka","Maharashtra","Tamil Nadu","Kerala","Delhi",
              "Gujarat","Rajasthan","Punjab","Uttar Pradesh","West Bengal"],

    "Population":[67500000,124000000,78000000,35600000,33000000,
                  71000000,81000000,31000000,240000000,100000000],

    "Literacy":[75.4,82.3,80.1,94.0,88.7,
                79.3,67.1,76.7,69.7,77.1],

    "Area":[191791,307713,130058,38863,1484,
            196024,342239,50362,243286,88752]
}

df = pd.DataFrame(data)

df
```

```python
df.to_csv("Indian_States.csv", index=False)
```

```python
df=pd.read_csv("Indian_States.csv")
```

```python
# Filter states having population greater than 80 million

filtered=df[df["Population"]>80000000]

filtered
```

```python
df["Region"]=["South","West","South","South","North",
              "West","North","North","North","East"]

df
```

```python
# Average literacy by region

group=df.groupby("Region")["Literacy"].mean()

group
```

```python
capital={
    "State":["Karnataka","Maharashtra","Tamil Nadu","Kerala",
             "Delhi","Gujarat","Rajasthan","Punjab",
             "Uttar Pradesh","West Bengal"],

    "Capital":["Bengaluru","Mumbai","Chennai","Thiruvananthapuram",
               "New Delhi","Gandhinagar","Jaipur","Chandigarh",
               "Lucknow","Kolkata"]
}

capital_df=pd.DataFrame(capital)
```

```python
merged=pd.merge(df,capital_df,on="State")
merged
```

```python
pivot=df.pivot_table(values="Population",
                     index="Region",
                     aggfunc="mean")
pivot
```

```python
merged.to_csv("Cleaned_Indian_States.csv", index=False)
```
