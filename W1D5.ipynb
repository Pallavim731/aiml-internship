```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create sample student dataset

data = {
    "Student_ID":[101,102,103,104,105,106,107,108,109,110],
    "Name":["Amit","Priya","Rahul","Sneha","Karan","Neha","Arjun","Pooja","Riya","Vikas"],
    "State":["Karnataka","Tamil Nadu","Kerala","Delhi","Maharashtra",
             "Kerala","Karnataka","Delhi","Tamil Nadu","Maharashtra"],
    "Course":["AI","AI","ML","AI","DS","ML","DS","AI","DS","ML"],
    "Marks":[85,92,76,88,95,81,79,90,84,87]
}

df = pd.DataFrame(data)

df
```

```python
# Display first five rows
df.head()

# Display information
df.info()

# Display statistical summary
df.describe()
```

```python
# Check missing values
df.isnull().sum()
```

```python
average = df["Marks"].mean()

print("Average Marks:", average)
```

```python
highest = df["Marks"].max()

print("Highest Marks:", highest)
```

```python
course_average = df.groupby("Course")["Marks"].mean()

print(course_average)
```

```python
plt.figure(figsize=(6,4))

plt.hist(df["Marks"], bins=5)

plt.title("Distribution of Student Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")

plt.show()
```

```python
df["Course"].value_counts().plot(kind="bar")

plt.title("Students in Each Course")

plt.xlabel("Course")

plt.ylabel("Count")

plt.show()
```

```python
df.to_csv("student_analysis.csv", index=False)
```
