```python
import pandas as pd

data = {
    "Student_ID": [101,102,103,104,105,106,107,108],
    "Name": ["Amit","Priya","Rahul","Sneha","Karan","Neha","Arjun","Pooja"],
    "State": ["Karnataka","Tamil Nadu","Kerala","Delhi","Maharashtra","Kerala","Karnataka","Delhi"],
    "Course": ["AI","AI","ML","AI","DS","ML","DS","AI"],
    "Marks": [85,92,76,88,95,81,79,90]
}

df = pd.DataFrame(data)

df.to_csv("students.csv", index=False)

print("students.csv created successfully!")
```

```python
numeric_columns = df.select_dtypes(include=['int64','float64'])

numeric_columns.hist(figsize=(10,5))

plt.show()
```

```python
corr = numeric_columns.corr()

plt.figure(figsize=(6,5))
plt.imshow(corr)
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")

plt.show()
```
