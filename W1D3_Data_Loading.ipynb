```python
# Import pandas library
import pandas as pd
```

```python
# Create sample Indian student dataset

data = {
    "Student_ID": [101,102,103,104,105,106,107,108],
    "Name": ["Amit","Priya","Rahul","Sneha","Karan","Neha","Arjun","Pooja"],
    "State": ["Karnataka","Tamil Nadu","Kerala","Delhi","Maharashtra","Kerala","Karnataka","Delhi"],
    "Course": ["AI","AI","ML","AI","DS","ML","DS","AI"],
    "Marks": [85,92,76,88,95,81,79,90]
}

df = pd.DataFrame(data)

df.to_csv("students.csv", index=False)

print("CSV Created Successfully!")
```

```python
# Load CSV file

df = pd.read_csv("students.csv")
```

```python
# Display first 10 rows

print(df.head(10))
```

```python
# Print number of rows and columns

print(df.shape)
```

```python
# Display datatype of each column

print(df.dtypes)
```

```python
# Students scoring more than 85 marks

high_marks = df[df["Marks"] > 85]

print(high_marks)
```

```python
# Average marks by course

group = df.groupby("Course")["Marks"].mean()

print(group)
```

```python
# Create faculty dataframe

faculty = pd.DataFrame({
    "Course":["AI","ML","DS"],
    "Faculty":["Anil","Ravi","Meena"]
})
```

```python
merged = pd.merge(df, faculty, on="Course")

print(merged)
```

```python
pivot = pd.pivot_table(
    df,
    values="Marks",
    index="State",
    aggfunc="mean"
)

print(pivot)
```

```python
df.to_csv("cleaned_students.csv", index=False)
```
